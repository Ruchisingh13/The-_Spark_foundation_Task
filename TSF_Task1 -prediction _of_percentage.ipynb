Task:- In this regression task 1 tried to predict the percentage of marks that a student is expected to score based upon the number of hours they studies.

**Importing libraries : Sikit Learn, Numpy Array, Pandas, Matplotilb**

```python
# Importing the required Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np   
```

## *Step 1 - Reading the data from source*

```python
# Reading data from remote link
Data = "http://bit.ly/w-data"
stud_data = pd.read_csv(Data)
print("Data imported successfully")

stud_data.head(10)
```

Step 2 - Input data Visualization

```python
# Plotting the distribution of scores
stud_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

*From the graph we can safely assume a positive linear relation between the number of hours studies and percentage of score.*

### *Step 3 - Data Preprocessing*

The step involved division of data into "attributes" (inputs) and "labels" (outputs).

```python
X = stud_data.iloc[:, :-1].values  
y = stud_data.iloc[:, 1].values 
X 
```

```python
y
```

### *Step 4 - Model Training*

Splitting the data into trining and testing sets, and training the algorithm.

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0) 
regressor = LinearRegression()
regressor.fit(X_train.reshape(-1,1),y_train)

print("training complete.")
```

### *step 5 - Plotting the Line of regression*
Now since our model is trained now, its the time of visualize the best-fit line of regression. 

```python
# Plotting the regression Line
line = regressor.coef_*X+regressor.intercept_

# Plotting for the test data
plt.scatter(X,y)  
plt.plot(X, line, color ='green');
plt.show()
```

### *Step 6 - Making predictions*

Now that we have trained our algorithm, its time to test the model by making some predictions.

For this we will use our test-set data

```python
# Testing data
print(X_test)

# Model Prediction
y_pred = regressor.predict(X_test)
```

### *Step 7 - Comparing Actual result to the predicted Model result*


```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
# Estimating training and test score
print("Training Score:-",regressor.score(X_train,y_train))
print("Test Score:-",regressor.score(X_test,y_test))
```

```python
# Plotting the Bar graph to depict the difference between the actual and predict value
color = ['red','green','blue','yellow']
df.plot(kind='bar',figsize= (7,7),color = color)
plt.grid(which='major',linewidth='0.5',color='red')
plt.grid(which='major',linewidth='0.5',color='blue')
plt.show()
```

```python
# Testing the model with our own data
hours = 9.25
test = np.array([hours])
test = test.reshape(-1,1)
own_pred = regressor.predict(test)
print("Number of hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

### *Step 8 - Evaluating the model*

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. Here different errors have been calculated to compare the model performance and predict the accuracy.   

```python
from sklearn import metrics  
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
print('Mean Squared Error:',metrics.mean_squared_error(y_test, y_pred)) 
print('Root Mean Squared Error:',np.sqrt(metrics.mean_absolute_error(y_test, y_pred))) 
print('R-2:',metrics.r2_score(y_test, y_pred)) 
```
